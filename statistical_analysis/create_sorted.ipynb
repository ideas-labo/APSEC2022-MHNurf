# create sorted table

```python
import pandas as pd
import ast

df2 = pd.DataFrame()
df = pd.read_excel('results.xlsx')

for project in ['tensorflow','pytorch','keras','mxnet','caffe']:
  for idx in range(df.shape[0]):
    data = []
    lst = df[project].iloc[idx]
    data = ast.literal_eval(lst)
    df2[df.iloc[idx,0]] = data

  df2.to_csv(f'/content/sorted_{project}.csv')
```
